**Demo Mean Shift Clustering algorithm menggunakan gambar 2D dan 3D**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift, estimate_bandwidth

# Generate synthetic 2D data
np.random.seed(0)
X_2d = np.random.randn(100, 2) * 2 + np.array([10, 10])

# Generate synthetic 3D data
X_3d = np.random.randn(100, 3) * 2 + np.array([10, 10, 10])

# Estimate bandwidth for 2D data
bandwidth_2d = estimate_bandwidth(X_2d, quantile=0.2, n_samples=len(X_2d))

# Estimate bandwidth for 3D data
bandwidth_3d = estimate_bandwidth(X_3d, quantile=0.2, n_samples=len(X_3d))

# Initialize Mean Shift with bandwidth for 2D data
ms_2d = MeanShift(bandwidth=bandwidth_2d, bin_seeding=True)

# Initialize Mean Shift with bandwidth for 3D data
ms_3d = MeanShift(bandwidth=bandwidth_3d, bin_seeding=True)

# Fit Mean Shift to the data
ms_2d.fit(X_2d)
ms_3d.fit(X_3d)

# Get cluster labels and cluster centers for 2D data
cluster_labels_2d = ms_2d.labels_
cluster_centers_2d = ms_2d.cluster_centers_

# Get cluster labels and cluster centers for 3D data
cluster_labels_3d = ms_3d.labels_
cluster_centers_3d = ms_3d.cluster_centers_

# Visualize clusters for 2D data
plt.figure(figsize=(8, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=cluster_labels_2d, cmap='viridis')
plt.scatter(cluster_centers_2d[:, 0], cluster_centers_2d[:, 1], c='red', marker='x', s=100, label='Cluster Centers')
plt.title('Mean Shift Clustering of 2D Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# Visualize clusters for 3D data
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c=cluster_labels_3d, cmap='viridis')
ax.scatter(cluster_centers_3d[:, 0], cluster_centers_3d[:, 1], cluster_centers_3d[:, 2],
           c='red', marker='x', s=100, label='Cluster Centers')
ax.set_title('Mean Shift Clustering of 3D Data')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.legend()
plt.show()
```
